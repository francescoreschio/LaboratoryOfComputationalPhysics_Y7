1. The MickeyMouse problem

a) Write a program that prints the numbers from 1 to 100. But for multiples of 3 print `Mickey` instead of the corresponding number and for the multiples of 5 print `Mouse`. For numbers which are multiples of both three and five print `MickeyMouse`

b) Put the result in a tuple and substitute `Mickey` with `Donald` and `Mouse` with `Duck`

```python


for i in range(1, 101):
    if (i % 5 == 0 and i % 3 == 0):
        print("MickeyMouse")
        List.append("MickeyMouse")
    elif i % 3 == 0:
        print("Mickey")
        List.append("Mickey")
    elif i % 5 == 0:
        print("Mouse")
        List.append("Mouse")
    else:
        print(i)
        

```

2\. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
def Swap(x, y):
    temp = x
    x = y
    y = temp
    return x, y

print(Swap(5, 4))

def SwapWithOutTemp(x, y):
    return y, x


print(SwapWithOutTemp(5, 4))
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
from math import sqrt
def Distance(Point1, Point2):

    xComp = Point1[0] - Point2[0]
    yComp = Point1[1] - Point2[1]
        
    return sqrt(xComp**2 + yComp**2)

x = (3, 0)
y = (0, 4)

Dist = Distance(x, y)
print(Dist)
```

4\. Counting letters

Write a program to calculate the number of times each character occurs in a given string *s*. Ignore differneces in capitalization

```python
string "questa è una stringa"
len(string
```

```python
def Count(string):
    stringLow = string.lower()
    
    counts = {}
    
    for i in stringLow:
        if i in counts:
            counts[i] += 1
            
        else:
            counts[i] = 1
        
    return counts
        
    
string = "Questa è una prova"
Dict = Count(string)

for key in Dict.keys():
	print(key, "appears ", Dict[key])
```
